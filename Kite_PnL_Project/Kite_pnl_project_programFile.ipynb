```python
import pandas as pd

# df = pd.read_csv(r"C:\Users\nanduji\Desktop\IotaAcadmey_DataAnalytic\python programing\Project file\python project files\Kite_Pnl_Project\orders (1).csv")
df = pd.read_csv("orders (1).csv")
df.head()
```

```python
# clean cloumns

# Filter add for MIS prouduct and Complete status
df = df[(df['Product'] == 'MIS') & (df['Status'] == 'COMPLETE')].copy()
df
# df['excuted_qty'] =df['Qty.'].str.split[0].astype(int)
```

```python
# Clean Qty column : take the part before'/' and covert to interger
# df['Qty.'] = df['Qty.'].str.split('/').str[0].astype(int) it sho
# Notes above code is give error becuase already have Qty is interger value then we change it str and after interger

df['Qty.'] = df['Qty.'].astype(str).str.split('/').str[0].astype(int)
df
```

```python
# Calculate Turnover(Quantity  * Price)

df['Turnover'] = df['Qty.'] * df['Avg. price']
df
```

```python
# Calculate Brokerage( capped at 20 or 0.03%)

df['Brokerage'] = df['Turnover'].apply(lambda x: min(x * 0.0003, 20))
df
```

```python
# STT/CTT = 0.025% on sell side only
df['STT/CTT'] = df.apply( lambda trade: trade['Turnover'] * 0.00025 if trade['Type'] == 'SELL' 
                          else 0, axis=1
)
df
```

```python
# calculate ETC(Exchange-Traded Commodity) Charges (0.0325) according to the NSE(National stock Exchange)
df['ETC'] = df['Turnover'] * 0.0000307
df

# Calcualte SEBI charge - according to Zerodha is per 10 cr =  0.0001%
df['SEBI'] =df['Turnover'] * 0.000001
df
```

```python
# Tax levied by the government on the services rendered. 18% of ( brokerage + SEBI charges + transaction charges)
# Calcualte Gst - (18% on brokerage + etc + sebi) 
df['GST'] =(df['Brokerage'] + df['ETC'] + df['SEBI']) * 0.18
df
```

```python
# Calculatem stamp Charges (accordinding Equity intraday treding =  0.003%)
df['Stamp charges']=df.apply(lambda data: data['Turnover'] * 0.00003 if data['Type'] == 'BUY' else 0, axis=1)
df
```

```python
# Total Charges 
df['Total charges'] = df[['Brokerage','STT/CTT','ETC','SEBI','GST','Stamp charges']].sum(axis=1).round(2)
df
```

```python

# we reasign indexing because of 
df = df.reset_index(drop=True)   #  assign back

df.index = range(1, len(df)+1)   # start index from 1

df = df.round(2)
```

```python


df_kite = pd.read_excel("kite_PnL_calculated_file.xlsx")

df_kite
```

```python
# GRoup by instrument and Type

table_2 = df_kite.groupby(['Instrument', 'Type'], as_index=False).agg({
    'Qty.': 'sum',
    'Turnover': 'sum',
    'Brokerage': 'sum',
    'STT/CTT': 'sum',
    'ETC': 'sum',
    'SEBI': 'sum',
    'GST': 'sum',
    'Stamp charges': 'sum',
    'Total charges': 'sum'
})
table_2
```

```python
# Weighted Avg Price = Total Turnover / Total Quantity
table_2['Avg. price'] = (table_2['Turnover'] / table_2['Qty.']).fillna(0)
table_2
```

```python
# reset indexing annd round off
table_2 = table_2.fillna(0)
table_2 = table_2.reset_index(drop=True)
table_2.index = range(1, len(table_2) + 1)
table_2 = table_2.round(2)
```

```python
# Reorder columns to match requirement
table_2 = table_2[['Instrument', 'Type', 'Qty.', 'Avg. price', 'Turnover',
                   'Brokerage', 'STT/CTT', 'ETC', 'SEBI', 'GST',
                   'Stamp charges', 'Total charges']].round(2)
table_2
```

```python
# --------------------------------------------------------------
```

```python
# Create separate dataframes for Buy and Sell to calculate PnL
```

```python

buy = table_2[table_2['Type'] == 'BUY'].groupby('Instrument')['Turnover'].sum()
sell = table_2[table_2['Type'] == 'SELL'].groupby('Instrument')['Turnover'].sum()


# total charges per stock
total_charges = table_2.groupby('Instrument')['Total charges'].sum()


# stock list
stocks = sorted(table_2['Instrument'].unique())


print(buy)
print("-----------------------")
print(sell)
print("-----------------------")
print('-----------------')
print(total_charges)
print('--------------------')
print(stocks)
```

```python
table_3= pd.DataFrame(index=stocks)
```

```python
print(table_3)
```

```python
# calculate Gross Pnl,Total charges adn Net Pnl
table_3['Gross PnL'] = sell.reindex(stocks, fill_value=0) - buy.reindex(stocks, fill_value=0)
table_3['Total charges'] = total_charges.reindex(stocks, fill_value=0)
table_3['Net PnL'] = table_3['Gross PnL'] - table_3['Total charges']
```

```python
table_3
```

```python
# Charges %
table_3['% Charges on Gross PnL'] = np.where(
    table_3['Gross PnL'] != 0,
    (table_3['Total charges'] / abs(table_3['Gross PnL'])) * 100,
    0
)
table_3
```

```python
# Reindexing  of index column 
table_3 = table_3.reset_index().rename(columns={'index': 'Instrument'})
table_3 = table_3.round(2)
```
