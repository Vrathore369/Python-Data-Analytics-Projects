```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta

from IPython.display import clear_output
```

```python
df = pd.read_excel('vaccine_data.xlsx')
df
```

```python
print(df.info())
```

##Vaccination Portal – Problem Statement

```python
# clen data
# strip()-remove extra space in string and lower() user for lowercase letters 
df['userid'] = df['userid'].str.strip().str.lower()

# it is conver string date type to actual date datatpye
df['date of 1st dose'] = pd.to_datetime(df['date of 1st dose'])
```

```python


# Take input
user_id = input("Enter user id: ").strip().lower()

# # user
user = df[df['userid'] == user_id]

if user.empty:
    print('User id not Found.')
else: 
    row = user.iloc[0]  # is use for index-based selection because we find value through index no.
    name = row['name']

    first_dose = row['1st dose status']      #
    second_dose = row['2nd dose status']  

     
    if first_dose == 'yes' and second_dose == 'yes':
        print(f"Hi {name}, Congratulations! You are fully vaccinated.")
    
    elif  first_dose  == 'no':
        print(f"Hi {name}, Please take your first dose.")
    
    else:
        first_date = row['date of 1st dose']
        next_date = first_date + pd.Timedelta(days=90)

        # Format date
        next_date_str = next_date.strftime('%d/%m/%Y')
        
        if pd.Timestamp.today() > next_date:
            print(f"Hi {name}, \nYou should have taken your 2nd dose on \n{next_date_str}.\nPlease take it as soon as possible.")
        else:
            print(f"Hi {name},  Take your 2nd dose on or after {next_date_str}.")
```

__Data Analysis__

```python
# 1.T1otal Users
Total_user = df['userid'].count()
print('Total User :',Total_user)
```

```python
# complted both vaccination  menmber list
competed_vaccianted = ((df['1st dose status'] == 'yes') & 
         (df['2nd dose status'] == 'yes')).sum()

print("Total fully vaccinated members:",competed_vaccianted )
```

```python
# only fist dose  comteted vaccination  menmber list
first_dose = (df['1st dose status'] == 'yes').count()
print('count of number of first dose:',first_dose)
```

```python

# user with without vaccination count
no_dose = df[
    (df['1st dose status'] == 'no') &
    (df['2nd dose status'] == 'no')
].shape[0]

print("No Dose:", no_dose)
```

```python
# no of user requried 2nd dose
today = pd.Timestamp("today")

overdue = df[
    (df['1st dose status'] == 'yes') &
    (df['2nd dose status'] == 'no') &
    ((today - df['date of 1st dose']).dt.days > 90)
].shape[0]

print("Second Dose Overdue:", overdue)
```

###Summary Table

```python
summary = pd.DataFrame({
    "Category": [
        "competed_vaccianted",
        "first_dose ",
        "No Dose",
        "Second Dose Overdue",
    ],
    "Count": [
        competed_vaccianted,
        first_dose,
        no_dose,
        overdue
    ]
})

summary
```

__Visulization__ 

```python
import matplotlib.pyplot as plt

```

```python
# Bar Chart 
plt.figure()
graph = summary.plot(kind='barh',
                  x='Category',
                  y='Count',
                  legend=False,
                  color=["green", "blue", "red", "orange"])

    
plt.title('Vaccination Summary Chart')
plt.xlabel('Category')
plt.ylabel('Count')

# Add data labels on top of each bar
for container in graph.containers:
    graph.bar_label(container, label_type='edge' ,padding=3)

plt.show()
```

```python
# pie Chart

plt.figure()

plt.pie(summary["Count"], labels=summary["Category"], autopct='%1.1f%%', colors=["green", "blue", "red", "orange"])

plt.title("Vaccination Distribution")

plt.show()
```
